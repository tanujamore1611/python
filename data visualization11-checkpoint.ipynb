```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([23,34,42,49,60,61,68,73,78,82])
y=np.array([25,29,33,37,44,47,57,62,67,71])
z=np.array([87,82,78,73,67,63,55,51,47,39])
```

```python
# line plot
```

```python
plt.title('line plot')
plt.xlabel('First value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker='o',mfc='red',color='purple',ls='--',ms=5);
plt.grid()

# markers=> >,<,^,H,o, *
```

```python
plt.title('line plot')
plt.xlabel('First value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,z,marker='o',mfc='yellow',color='purple',ls='--',ms=5);
plt.grid()
```

```python
plt.title('line plot')
plt.xlabel('First value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker='o',mfc='red',color='purple',ls='--',ms=5);
plt.plot(x,z,marker='H',mfc='yellow',color='purple',ls='--',ms=5);
plt.grid()
```

```python
#subplot
plt.figure(figsize=(16,9))

plt.subplot(2,2,1)
plt.title('line plot')
plt.xlabel('First value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker='o',mfc='red',color='purple',ls='--',ms=5);
plt.grid()

plt.subplot(2,2,2)
plt.title('line plot')
plt.xlabel('First value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,z,marker='H',mfc='yellow',color='purple',ls='--',ms=5);
plt.grid()
```

```python
courses=['java','python','kotlin','dart','R']
students=[70,60,50,40,30]
```

```python
plt.title('Enrollnment')
plt.xlabel('Courses')
plt.ylabel('students')
plt.bar(courses,students,color='gray',edgecolor='purple',width=0.5,linewidth=3,alpha=0.9);
plt.grid()
```
